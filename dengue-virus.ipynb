```python
# Asn or Asp / B	AAU, AAC; GAU, GAC
# Gln or Glu / Z	CAA, DAG; GAA, GAG
# START	AUG -  stupid as this is just a MET

codonTable = """Ala / A	GCU, GCC, GCA, GCG
Ile / I	AUU, AUC, AUA
Arg / R	CGU, CGC, CGA, CGG; AGA, AGG
Leu / L	CUU, CUC, CUA, CUG; UUA, UUG
Asn / N	AAU, AAC
Lys / K	AAA, AAG
Asp / D	GAU, GAC
Met / M	AUG
Phe / F	UUU, UUC
Cys / C	UGU, UGC
Pro / P	CCU, CCC, CCA, CCG
Gln / Q	CAA, CAG
Ser / S	UCU, UCC, UCA, UCG; AGU, AGC
Glu / E	GAA, GAG
Thr / T	ACU, ACC, ACA, ACG
Trp / W	UGG
Gly / G	GGU, GGC, GGA, GGG
Tyr / Y	UAU, UAC
His / H	CAU, CAC
Val / V	GUU, GUC, GUA, GUG
STOP	UAA, UGA, UAG""" # you can refer here: https://en.wikipedia.org/wiki/DNA_and_RNA_codon_tables
# note both Met and START are AUG - deal with it, Krish

dec = {}  #dictionary to use to decode DNA
for i in codonTable.split('\n'):
    acid,codons = i.split("\x09") # split based on tabs
    if '/' in acid :
        acid = acid.split("/")[-1].strip()
    acid = acid.replace("STOP", "#")
    codons = codons.replace(",", "").replace(";", "").lower()  # lower case it after removing the , and ;
    codons = codons.replace("u", "t").split(" ")   # put it in the form like A [''] 
    for codon in codons:
        if codon in dec:
            print("duplicate", codon)
        dec[codon] = acid

          
#dec 
```

```python
polypeptides = [] # chain of amino acids, proteins are a group of these

for readingFrame in range(3):
    
    for i in range(readingFrame, len(dengue)-3, 3):
        polypeptides.append(dec[dengue[i:i+3]])
        
polypeptides = ''.join(polypeptides)
```
